```python
import pandas as pd

# Load the dataset
df = pd.read_csv('top_rated_movies.csv', index_col=0)

# Reset the index to turn the index into a regular column
df.reset_index(inplace=True)

# Rename the new column to 'id'
df.rename(columns={'index': 'id'}, inplace=True)

# Save the updated dataset to a new CSV file
df.to_csv('updated_dataset.csv', index=False)
data = pd.read_csv('updated_dataset.csv')
data
```

```python
data.head(10)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 10000,stop_words = 'english')
cv
```

```python
data['tags'] = data['overview'] + data['genre_ids']
```

```python
new_data = data.drop(columns=['overview','genre_ids'])
```

```python
vector = cv.fit_transform(new_data['tags'].values.astype('U')).toarray()
vector
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarilty = cosine_similarity(vector)
similarilty
```

```python
new_data[new_data['title']=="The Godfather"]
```

```python
new_data.info()
```

```python
def recommand(data):
     movie_index = new_data[new_data['title'] == data].index[0]

     distance = sorted(list(enumerate(similarilty[movie_index])) ,reverse = True,key = lambda vector:vector[1])
     for i in distance[0:5]:
         print(new_data.iloc[i[0]].title)
```

```python
recommand("Iron Man")
```

```python
import pickle
pickle.dump(new_data,open('movies_list.pkl','wb'))
pickle.dump(similarilty,open('similarilty.pkl','wb'))
pickle.load(open('movies_list.pkl','rb'))
```
